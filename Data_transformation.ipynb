```python
import pandas as pd
```

```python
df = pd.read_csv(r"/content/My dataset.csv")
df
```

```python
df = df.drop_duplicates()
df
```

```python
object_columns = df.select_dtypes(include=['object']).columns
df[object_columns] = df[object_columns].astype('category')
df
```

```python
df['Age'] = df['Age'].astype(int)
df
```

```python
df['BodyMassIndex'] = df['BMI'].astype(float)
df
```

```python
df = pd.read_csv('My dataset.csv',names = ['Num Pregnancies','Blood Glucose','Blood pressure','SkinThickness','InsulinLevel','BodyMassIndex','PedigreeScore','YearsOld','DiabetesOutcome'],header = 1)
```

```python
df.head()
```

```python
df['GlucoseToBloodPressureRatio'] = df['Blood Glucose'] / df['Blood pressure']
df

```

```python
df['DiabetesRiskScore'] = (df['Blood Glucose'] * 0.4) + (df['BodyMassIndex'] * 0.3) - (df['YearsOld'] * 0.1)
df
```

```python
df['TotalInsulin'] = df['InsulinLevel'] * df['Num Pregnancies']
df
```

```python
mean_glucose_by_age = df.groupby('YearsOld')['Blood Glucose'].mean()
df
```

```python
df = df [['Num Pregnancies','Blood Glucose','Blood pressure','GlucoseToBloodPressureRatio','SkinThickness','InsulinLevel','BodyMassIndex','DiabetesRiskScore','TotalInsulin','PedigreeScore','YearsOld',	'DiabetesOutcome']]
df
```

```python
median_skin_thickness = df['SkinThickness'].median()
df['SkinThickness'].fillna(median_skin_thickness, inplace=True)
df
```

```python
df.to_csv('my_new_data.csv',index=False)
```
